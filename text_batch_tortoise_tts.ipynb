# process sentences

```python
from tqdm.auto import tqdm

curr_seed = 42

save_path = "/content/drive/MyDrive/AI/SOUND/tortoise_batch/"
preset = "high_quality"

voice = 'train_atkins'
voice_samples, conditioning_latents = load_voice(voice)

cpt = 0
for sentence in tqdm(final_sentence_list):
  gen = tts.tts_with_preset(sentence, voice_samples=voice_samples, conditioning_latents=conditioning_latents, preset=preset, use_deterministic_seed=curr_seed)

  filename = save_path + f"speech_track_{cpt}.wav"
  cpt = cpt + 1
  
  torchaudio.save(filename, gen.squeeze(0).cpu(), 24000)
  # IPython.display.Audio(filename)
```
